# Função densidade de probabilidade (PDF) e distribuição acumulada (CDF)

Em teoria das probabilidades e estatística, a função densidade de probabilidade (FDP), ou densidade de uma variável aleatória contínua, é uma função que descreve a verossimilhança de uma variável aleatória tomar um valor dado. A probabilidade da variável aleatória cair em uma faixa particular é dada pela integral da densidade dessa variável sobre tal faixa - isto é, é dada pela área abaixo da função densidade mas acima do eixo horizontal e entre o menor e o maior valor dessa faixa. A função densidade de probabilidade é não negativa sempre, e sua integral sobre todo o espaço é igual a um. A função densidade pode ser obtida a partir da função distribuição acumulada a partir da operação de derivação (quando esta é derivável).

Essa fórmula representa a probabilidade de uma variável aleatória contínua ( X ) assumir valores dentro do intervalo ([a, b]), com ( f(x) ) sendo a função densidade de probabilidade

\begin{align*}
P[a \leq X \leq b] = \int_{a}^{b} f(x) \, dx
\end{align*}

Isso define a função de distribuição acumulada (CDF) para uma variável aleatória (X). 

\begin{align*}
F_X(x) = P[X \leq x] = \int_{-\infty}^{x} f(X) \, dX
\end{align*}



```python
import numpy as np
from scipy.integrate import quad

# Definição da função de densidade de probabilidade f(x)
# Exemplo: distribuição normal com média 0 e desvio padrão 1
def f(x):
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-x**2 / 2)

# Limites do intervalo [a, b]
a = -1
b = 1

# Cálculo da probabilidade no intervalo [a, b]
probabilidade, _ = quad(f, a, b)

print(f"A probabilidade de X estar no intervalo [{a}, {b}] é: {probabilidade:.4f}")
```

# Teste de Hipótese

Realizamos uma afirmação sobre um parametro da população, de acordo com a afirmação temos duas hipóteses:

- Hipótese Nula ($H_0$) - Representa a Ausência de evento ou mudança / É a afirmação sobre o valor de um parametro da população

- Hipótese Alternativa ($H_1$) - Representa a presença de um efeito ou mudança / hipótese que aceitamos caso $H_0$ refutada




- **Nível de significância** $(\alpha)$: Probabilidade máxima permitida de cometer um erro do Tipo I (rejeitar $H_0$ quando ela é verdadeira). Normalmente definido como $0.05$ (5%).

- **Potência desejada** $(1 - \beta)$: Probabilidade de rejeitar $H_0$ quando $H_1$ é verdadeira. Geralmente fixado em $80%$ ou $90%$.

$$
\begin{array}{|l|l|l|}
\hline
                    & \textbf{Verdadeiro}                & \textbf{Falso}                      \\ \hline
\text{Não rejeita a } H_0 & \text{Correta}                 & \text{Erro tipo II } (\beta)         \\ \hline
\text{Rejeita } H_0       & \text{Erro tipo I } (\alpha)   & \text{Correta}                      \\ \hline
\end{array}
$$

Erro do Tipo I ($\alpha$) é quando rejeita $H_0$ quando ela é verdadeira

Erro do tipo II ($\beta$) é quando não rejeita $H_0$ quando $H_1$ é verdadeira

- Se aumenta o **erro tipo I** diminui o **erro tipo II** 
- Se aumenta o **erro tipo II** diminui o **erro tipo I**

----

**p-value** é a probalidade de se observar valores extremos a estatistica observada

- Se $p-value < \alpha$ então Rejeita-se a $H_0$
- Se $p-value >= \alpha$ então Não rejeita-se a $H_0$



### Etapas para um teste de Hipótese:

1. Formulação das hipóteses
2. Escolha do Nível de Siginificância ($\alpha$)
3. Seleção do Teste Apropriado
4. Cálculo da Estatística de Teste
5. Determinar o $p-value$
6. Comparação com $\alpha$
7. Tomada de Decisão

```python
import scipy.stats as stats
import numpy as np

# Dados: duas amostras independentes
grupo_1 = [20, 21, 19, 22, 20, 23, 24, 22, 21, 20]
grupo_2 = [30, 28, 29, 31, 32, 30, 33, 29, 28, 30]

# 1. Formulação das hipóteses
# H0: A média dos dois grupos é igual (não há diferença significativa)
# H1: A média dos dois grupos é diferente (há diferença significativa)

# 2. Escolha do Nível de Siginificância ($\alpha$)
alpha = 0.05

# 3. Seleção do Teste Apropriado
# Calcula o teste t para as médias de duas amostras independentes de pontuações.
# Este é um teste bilateral para a hipótese nula de que 2 amostras independentes têm valores médios (esperados) idênticos.
# Este teste assume, por padrão, que as populações têm variâncias idênticas.

# 4. Cálculo da Estatística de Teste
stat, p_value = stats.ttest_ind(grupo_1, grupo_2)

# 5. Determinar o $p-value$
print("Estatística t:", stat)
print("Valor p:", p_value)


# 6. Comparação com $\alpha$ e # 7. Tomada de Decisão
if p_value < alpha:
    print("Rejeitamos a hipótese nula. Há diferença significativa entre os grupos.")
else:
    print("Não rejeitamos a hipótese nula. Não há diferença significativa entre os grupos.")
```

## Distribuição Normal

A distribuição normal, também conhecida como distribuição Gaussiana, é uma das distribuições de probabilidade mais importantes e amplamente usadas em estatística e ciência. 

Ela é caracterizada por sua forma de sino, simétrica em relação à média.

Uma variável aleatória $X$ segue uma distribuição normal se sua função densidade de probabilidade for dada por:

$$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$

Onde:
- $\mu$ é o valor esperado (média).
- $\sigma^2$ é a variância.
- $\sigma$ é o desvio padrão.

Valor Esperança:
$$E(X) = \mu$$

Variância:
$$\text{Var}(X) = \sigma^2$$

**Propriedades principais**:

1. A média ($\mu$) é o centro da curva e determina a posição.
2. O desvio padrão ($\sigma$) controla a largura da curva: quanto maior $\sigma$, mais "achatada" é a curva.
3. A distribuição é simétrica em torno da média.
4. Aproximadamente 68%, 95%, e 99.7% dos valores caem dentro de $\text{1}\sigma$, $\text{2}\sigma$, e $\text{3}\sigma$, da média, respectivamente (regra empírica).

```python
from scipy.stats import norm
import seaborn as sns

# Definir os parâmetros da distribuição normal (média e desvio padrão)
mu = 0  # média
sigma = 1  # desvio padrão

x = np.linspace(-4, 4, 1000)

pdf = norm.pdf(x=x, loc=mu, scale=sigma)
print(f"O valor da função de densidade de probabilidade no ponto x=1 é: {pdf[(np.abs(x - 0)).argmin()]:.4f}")

sns.lineplot(x=x, y=pdf, label="f(x)", color="blue")
plt.show()
```

**Z-score**

Na distribuição normal, o $Z$ é um valor padronizado, também chamado de score $Z$ ou $Z-score$, que representa a posição de um dado em relação à média da distribuição, medido em desvios padrão.

Ele permite comparar valores em diferentes distribuições normais, transformando-os para uma escala comum:

$$ Z = \frac{X - \mu}{\sigma}$$

Onde:
- Z: O valor z-score
- X: O valor da observação(dado)
- $\mu$: A média da distribuição
- $\sigma$: O desvio padrão da distribuição

Quando:
- $z-score = 0$ está na média
- $z-score > 0$ valor acima da média
- $z-score < 0$ valor abaixa da média


Exemplo: Dado $\mu$ = 300 e $\sigma$ = 50, Probalidade de X > 362

```python
X = 362
mu = 300
sigma = 50

Z = (X - mu) / sigma
print(f"Z-score: {Z:.4f}")

probabilidade = norm.cdf(Z)
print(f"Probabilidade acumulada de X <= {X}: {probabilidade:.4f}")
print(f'Logo a probalidade de X > {X} é: {1 - probabilidade:.4f}')
```

**Curtuse**

A curtose é uma medida estatística que descreve a forma da cauda de uma distribuição em relação a uma distribuição normal. 
Ela indica o grau em que os valores extremos (outliers) estão presentes.

Distribuição mesocúrtica, dizemos que não há excesso de curtose e esse é o caso da já discutida distribuição normal.

Já em uma distribuição leptocúrtica, a distribuição tem excesso positivo de curtose, com caudas mais pesadas. Em outras palavras, há mais outliers em uma distribuição leptocúrtica do que esperaríamos em uma distribuição normal. Por exemplo, a distribuição t com poucos graus de liberdade é leptocúrtica, pois ela tem caudas mais pesadas que uma distribuição normal.

Por fim, em uma distribuição platicúrtica, a distribuição tem excesso negativo de curtose, em comparação com a distribuição normal. Um exemplo é a distribuição uniforme, onde os diferente valores são equiprováveis.

- Curtuse maior que 3 então **Leptocúrtica**
- Curtuse igual a 3 então **Mesocúrtica**
- Curtuse menor que 3 então **Platicúrtica**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats


# Configurando o número de pontos
np.random.seed(42)
n = 10000  # Número de pontos

# 1. Mesocúrtica (Distribuição Normal)
mesocurtic = np.random.normal(loc=0, scale=1, size=n)
kur_mesocurtic= stats.kurtosis(mesocurtic, fisher=True)

# 2. Leptocúrtica (Distribuição com caudas mais pesadas, ex: distribuição Laplace)
leptokurtic = np.random.laplace(loc=0, scale=1, size=n)
kur_leptokurtic= stats.kurtosis(leptokurtic, fisher=True)

# 3. Platicúrtica (Distribuição achatada, ex: distribuição uniforme)
platicurtic = np.random.uniform(low=-np.sqrt(3), high=np.sqrt(3), size=n)
kur_platicurtic= stats.kurtosis(platicurtic, fisher=True)

# Criando os gráficos
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

# Mesocúrtica (Normal)
axes[0].hist(mesocurtic, bins=50, color='blue', alpha=0.7, density=True, label=f'Mesocúrtica {kur_mesocurtic:.2f}') 
axes[0].set_title("Mesocúrtica (Normal)")
x = np.linspace(-4, 4, 1000)
axes[0].plot(x, stats.norm.pdf(x, loc=0, scale=1), 'r', lw=2, label=f'Mesocúrtica {stats.kurtosis(stats.norm.pdf(x, loc=0, scale=1), fisher=True):.2f}') # Curva normal
axes[0].legend()

# Leptocúrtica (Laplace)
axes[1].hist(leptokurtic, bins=50, color='green', alpha=0.7, density=True, label=f'Leptocúrtica {kur_leptokurtic:.2f}')
axes[1].set_title("Leptocúrtica (Laplace)")
axes[1].legend()

# Platicúrtica (Uniforme)
axes[2].hist(platicurtic, bins=50, color='orange', alpha=0.7, density=True, label=f'Platicúrtica {kur_platicurtic:.2f}')
axes[2].set_title("Platicúrtica (Uniforme)")
axes[2].legend()

plt.tight_layout()
plt.show()

print(f'kurtosis da mesocúrtica: {kur_mesocurtic:.4f}')
print(f'kurtosis da leptocúrtica: {kur_leptokurtic:.4f}')
print(f'kurtosis da platicúrtica: {kur_platicurtic:.4f}')
```

## Distribuição Uniforme



A distribuição uniforme descreve uma situação em que todos os resultados dentro de um intervalo têm a mesma probabilidade de ocorrer.

 Ela pode ser contínua ou discreta, dependendo do conjunto de valores possíveis.

Definição da função densidade de probabilidade da Distribuição Uniforme (Contínua):

$$f(x) =
\begin{cases} 
\frac{1}{b-a}, & a \leq x \leq b \\
0, & \text{caso contrário}
\end{cases}
$$

Valor Esperança:

$$
E(X) = \frac{a + b}{2}$$

Variância:

$$\text{Var}(X) = \frac{(b-a)^2}{12}$$

Onde:
- $a$: limite inferior da distribuição
- $b$: limite superior da distribuição
- $x$: valor da variável aleatória.

Os valores no intervalo $[a,b]$ têm a mesma probalidade.


```python
import numpy as np
import matplotlib.pyplot as plt

# Configuração dos parâmetros da distribuição uniforme
a = 0  # Limite inferior
b = 10  # Limite superior
n = 10000  # Número de pontos

# Geração de valores aleatórios uniformes
uniform_values = np.random.uniform(low=a, high=b, size=n)

# Criando o gráfico
plt.figure(figsize=(10, 6))

# Histograma da distribuição uniforme
plt.hist(uniform_values, bins=50, color='blue', alpha=0.7, edgecolor='black', density=True)
sns.lineplot(x=uniform_values, y=0.11, label="f(x)", color="red")
plt.title("Distribuição Uniforme Contínua")
plt.xlabel("Valor")
plt.ylabel("Frequência")
plt.grid(alpha=0.3)

# Exibição do gráfico
plt.show()
```

### Exemplo: A probalidade de um ponte escolhido ao acaso do segmento de reta [0,2] estar entra 1 e 3/2 é igual.

```python
# Limites do intervalo
a, b = 0, 2  # [0, 2]

# Subintervalo entre 1 e 3/2
x1, x2 = 1, 3/2

# Cálculo da probabilidade
if x1 < a or x2 > b:
    raise ValueError("O intervalo [x1, x2] não está contido no intervalo [a, b].")
else:
    probabilidade = (x2 - x1) / (b - a)
    # Exibindo o resultado
    print(f"A probabilidade de um ponto estar entre {x1} e {x2} é: {probabilidade:.2f}")    
```

## Distribuição Bernoulli

A distribuição Bernoulli é uma das distribuições de probabilidade mais simples e fundamentais, usada para modelar experimentos que possuem apenas dois resultados possíveis: 
- sucesso (1) 
- falha (0).

Seja $𝑋$ uma variável aleatória que segue uma distribuição Bernoulli a função de massa de probabilidade é dada por:
$$ P(X = x) = \begin{cases} p, & \text{se } x = 1 \\ 1-p, & \text{se } x = 0 \end{cases} $$

Também pode ser escrito como:
$$ P(X = x) = p^x(1-p)^{1-x}$$

onde:
- $p$: Probalidade de sucesso ($0 \leq p \leq 1$)
- $1-p$: Probalidade de falha.
- $x \in [0,1]$: Resultado do experimento.

**Valor Esperado**:
$$E(X) = p$$

**Variância**:
$$\text{Var}(X) = p(1-p)$$

A distribuição Bernoulli é usada para modelar situações como:
- O resultado de lançar uma moeda (cara ou coroa).
- Respostas binárias a uma pergunta ("sim" ou "não").
- Sucesso ou falha em um experimento único.

## Distribuição Binomial

A distribuição binomial é uma distribuição de probabilidade que modela o número de sucessos em $𝑛$ tentativas independentes de um experimento, onde cada tentativa tem apenas dois possíveis resultados: 
- sucesso (1) 
- falha (0)

É amplamente utilizada em estatística para analisar processos binários, como lançar uma moeda $𝑛$ vezes.

Seja $𝑋$ o número de sucessos em $𝑛$ tentativas, com probabilidade de sucesso $𝑝$ em cada tentativa.

A função de massa de probabilidade (FMP) é dada por:

$$ P(k) = \binom{n}{k} p^k (1-p)^{n-k} $$ 

Onde:  
- $k$: é o número de sucessos ($k \in \{0,1,2, \dots, n\}$ ) 
- $p$: é a probabilidade de sucesso em um unica tentativa
- $1-p$: Probalidade de falha em um unica tentativa
- $n$: Número total de tentativas.
- $\binom{n}{k}$: Coeficiente Binomial, calculado como $\binom{n}{k} = \frac{n!}{k!(n-k)!}$

**Valor Esperado**:
$$E(X) = np$$

**Variância**:
$$\text{Var}(X) = np(1-p)$$

A distribuição Bernoulli é usada para modelar situações como:
- O resultado de lançar uma moeda (cara ou coroa).
- Respostas binárias a uma pergunta ("sim" ou "não").
- Sucesso ou falha em um experimento único.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parâmetros da distribuição binomial
n = 10   # Número de experimentos (tentativas)
p = 0.5  # Probabilidade de sucesso
tamanho_amostra = 10000  # Número de amostras, Total de valores simulados.

# Gerando valores da distribuição binomial
dados = np.random.binomial(n=n, p=p, size=tamanho_amostra)

# Frequências relativas
valores, frequencias = np.unique(dados, return_counts=True)
frequencias_relativas = frequencias / tamanho_amostra

# Criando o gráfico
plt.figure(figsize=(10, 6))
plt.bar(valores, frequencias_relativas, color='blue', alpha=0.7, edgecolor='black')
plt.title(f"Distribuição Binomial (n={n}, p={p})")
plt.xlabel("Número de sucessos")
plt.ylabel("Frequência relativa")
plt.grid(axis='y', alpha=0.3)

# Exibindo o gráfico
plt.show()
```

## Distribuição Geométrica

A distribuição geométrica é uma distribuição de probabilidade discreta que modela o número de tentativas necessárias para obter o primeiro sucesso em uma sequência de experimentos independentes, onde cada experimento tem a mesma probabilidade de sucesso $𝑝$.

Seja $𝑋$ uma variável aleatória que representa o número de tentativas até o primeiro sucesso.

Sua função de massa de probabilidade (FMP) é:

$$ P(X = k) = p(1-p)^{k-1}  $$

Onde:
- $k$: Número de tentativas até o primeiro sucesso ($k \in \{1,2,3, \dots\}$)
- $p$: Probalidade de sucesso em cada tentativas ($0 \leq p \leq 1$).
- $1-p$: Probalidade de falha em cada tentativa.

**Valor Esperado**:
$$ E(X) = \frac{1}{p} $$

**Variância**:
$$ \text{Var}(X) = \frac{1-p}{p^2} $$

Exemplo Número de lançamento de um dado até obter um determinado número


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

# Parâmetros da distribuição geométrica
p = 0.4  # Probabilidade de sucesso
tamanho_amostra = 10000  # Número de amostras

# Gerando valores da distribuição geométrica
dados = np.random.geometric(p, tamanho_amostra)

# Frequência dos valores
valores, frequencias = np.unique(dados, return_counts=True)

# Criando o gráfico
plt.figure(figsize=(10, 6))
plt.bar(valores, frequencias / tamanho_amostra, color='blue', alpha=0.7, edgecolor='black')
plt.title
```

## Distribuição Poisson

A **distribuição de Poisson** é uma distribuição de probabilidade discreta que descreve o número de eventos que ocorrem em um intervalo fixo de tempo ou espaço, dado que os eventos ocorrem independentemente e a uma taxa constante.

Seja $𝑋$ o número de eventos em um intervalo fixo.

A função de massa de probabilidade (FMP) é dada por:

$$ P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!} $$

Onde:
- $k$: Número de eventos ($k \in \{0,1,2,\dots,\}$)
- $\lambda$: Taxa média de ocorrência dos eventos no intervalo
- $e$: A constante exponencial

**Valor Esperado**:

$$ E(X) = \lambda $$

**Variância**:

$$ \text{Var}(X) = \lambda $$


**Condições de Aplicação**:
- Os Eventos devem ocorrer de forma independente.
- A taxa média ($\lambda$) deve ser constante no intervalo analisado.
- Dois eventos não podem ocorrer exatamente ao mesmo tempo (distribuição discreta).




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parâmetro da distribuição de Poisson
lambda_poisson = 4  # Taxa média de ocorrências no intervalo

# Gerando valores aleatórios de Poisson
tamanho_amostra = 10000
dados_poisson = np.random.poisson(lam=lambda_poisson, size=tamanho_amostra)

# Frequências dos valores
valores, frequencias = np.unique(dados_poisson, return_counts=True)

# Criando o gráfico
plt.figure(figsize=(10, 6))
plt.bar(valores, frequencias / tamanho_amostra, color='blue', alpha=0.7, edgecolor='black')
plt.title(f'Distribuição de Poisson ($\\lambda$ = {lambda_poisson})')
plt.xlabel('Número de eventos (k)')
plt.ylabel('Frequência relativa')
plt.grid(alpha=0.3)

# Exibindo o gráfico
plt.show()
```

(Magalhães, 2004) Uma indústria de tintas recebe pedidos de seus vendedores 
através de fax, telefone e internet. A taxa média é de 5 pedidos por hora.

- Qual a probabilidade da indústria receber mais de dois pedidos por hora? 
Digamos que, no horário do almoço, a indústria fica impossibilitada de atender a 
mais de dois pedidos por hora. Você acha que deveria aumentar o nº de atendentes 
nesse período? 

```python
#Para receber mais de 2 pedidos por hora, temos P(x>2) = 1 - P(x<=2)
from scipy.stats import poisson

# Parâmetro da distribuição de Poisson
lambda_poisson = 5 # Taxa média 5 pedidos por hora

# Valor desejado (número de pedidos)
k = [0,1,2]
list_prob = []
for i in k:
    probabilidade = poisson.pmf(i, lambda_poisson)
    print(f"A probabilidade de receber {i} pedidos em 1 hora é: {probabilidade:.6f}")
    list_prob.append(probabilidade)

# A probabilidade de receber mais de 2 pedidos é 1 - P(x<=2)
probabilidade_mais_2 = 1 - sum(list_prob)
print(f"A probabilidade de receber mais de 2 pedidos em 1 hora é: {probabilidade_mais_2:.6f}")


```
