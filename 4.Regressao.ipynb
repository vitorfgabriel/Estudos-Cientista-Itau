   √â a parte que incorpora as vari√°veis explicativas ou preditores. Nela, usamos uma combina√ß√£o linear dos preditores:
   
   $$
   \eta = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_k x_k
   $$
   
   onde $\eta$ √© chamado de **predictor linear** e os coeficientes $\beta_i$ s√£o os par√¢metros a serem estimados.

3. **Fun√ß√£o de Liga√ß√£o (Link Function):**  

   Essa fun√ß√£o conecta o preditor linear √† m√©dia condicional da vari√°vel resposta. Ou seja, define uma rela√ß√£o entre $\mu = E(Y)$ e o preditor $\eta$:
   
   $$
   g(\mu)= \eta
   $$
   
   A escolha da fun√ß√£o de liga√ß√£o $g(\cdot)$ depende da natureza dos dados. Por exemplo:
   
   - **Identidade:** Para dados gaussianos, onde $g(\mu)=\mu$.
   - **Logito:** Para dados bin√°rios, usando $g(\mu)=\log\left(\frac{\mu}{1-\mu}\right)$.
   - **Logaritmo:** Para contagens (regress√£o Poisson), onde $g(\mu)=\log(\mu)$.

---


### Estrutura Matem√°tica de um GLM

Em resumo, o modelo assume que cada observa√ß√£o $y_i$ tem uma distribui√ß√£o da fam√≠lia exponencial, com a fun√ß√£o de densidade ou massa escrita genericamente como

$$
f(y_i; \theta_i, \phi) = \exp\left\{\frac{y_i \theta_i - b(\theta_i)}{\phi} + c(y_i,\phi)\right\},
$$

onde:
- $\theta_i$ √© o par√¢metro natural (ou can√¥nico) do modelo, que, em muitos casos, est√° diretamente relacionado √† m√©dia.
- $\phi$  √© o par√¢metro de dispers√£o (ou escala), presente em muitas distribui√ß√µes para ajustar a variabilidade dos dados.
- $b(\theta)$ √© uma fun√ß√£o conhecida como cumulante, que garante a normaliza√ß√£o da fun√ß√£o.
- $c(y,\phi)$ √© uma fun√ß√£o que depende somente de $y$ e $\phi$ e, em geral, serve para normalizar a distribui√ß√£o.


$$
f_X(x|\theta) = h(x)\exp[\eta(\theta).T(x) - A(\theta)]
$$

A fun√ß√£o de liga√ß√£o √© ent√£o usada para relacionar o par√¢metro de localiza√ß√£o (muitas vezes, a m√©dia $\mu_i$) ao preditor linear, atrav√©s de

$$
g(\mu_i) = \beta_0 + \beta_1 x_{i1} + \cdots + \beta_k x_{ik}.
$$


### Valor Esperado

O valor esperado (m√©dia) de $y_i$ √© dado por:

$$
E(y_i) = \mu_i = b'(\theta_i)
$$

Ou seja, a m√©dia dos dados est√° diretamente relacionada √† primeira derivada de $b(\theta)$ em rela√ß√£o a $\theta$.

### Vari√¢ncia

A vari√¢ncia de $y_i$ √© expressa como:

$$
\operatorname{Var}(y_i) = \phi \, b''(\theta_i)
$$

Isso mostra que a vari√¢ncia √© proporcional √† segunda derivada de $b(\theta)$ e ajustada pelo par√¢metro de dispers√£o $\phi$.


### Comparativos

| Tipo de GLM            | Distribui√ß√£o da Resposta | Fun√ß√£o de Liga√ß√£o       | Aplica√ß√£o Comum                             |
|------------------------|--------------------------|-------------------------|---------------------------------------------|
| **Regress√£o Linear**   | Normal                   | Identidade: $$g(\mu) = \mu$$ | Previs√£o de valores cont√≠nuos               |
| **Regress√£o Log√≠stica**| Binomial                 | Logito: $$g(\mu) = \log\left(\frac{\mu}{1-\mu}\right)$$ | Modelagem de probabilidades (classifica√ß√£o) |
| **Regress√£o Poisson**  | Poisson                  | Logaritmo: $$g(\mu) = \log(\mu)$$ | Contagem de eventos (ex.: n√∫mero de ocorr√™ncias) |

```
                         ‚îÇ

               (Œ∑ = Œ≤0 + Œ≤1x1 + ‚Ä¶ + Œ≤kxk)

```

---

### Por Que Utilizar GLMs?

- **Flexibilidade:** Permite modelar diferentes tipos de vari√°veis resposta (cont√≠nuas, bin√°rias, de contagem, etc.).
- **Integra√ß√£o com a Fam√≠lia Exponencial:** A abordagem unifica modelos diferentes sob uma mesma estrutura, facilitando t√©cnicas de estima√ß√£o, diagn√≥stico e valida√ß√£o.
- **Interpreta√ß√£o:** Embora a interpreta√ß√£o dos coeficientes dependa da fun√ß√£o de liga√ß√£o escolhida, os GLMs fornecem uma forma consistente de analisar a rela√ß√£o entre preditores e resposta.

---

### Reflex√£o e Aplica√ß√µes

Pense nos GLMs como uma poderosa caixa de ferramentas para lidar com uma variedade de problemas de modelagem. Por exemplo, se voc√™ estiver trabalhando com dados de incid√™ncia de uma doen√ßa em diferentes regi√µes, a regress√£o Poisson pode ser o modelo ideal. Ou, se a quest√£o for prever o sucesso ou fracasso de uma campanha, a regress√£o log√≠stica se encaixa perfeitamente.

Voc√™ j√° experimentou utilizar GLMs em algum projeto? Como voc√™ lida com a escolha da fun√ß√£o de liga√ß√£o ou com a interpreta√ß√£o dos coeficientes em contextos n√£o-lineares? Se houver interesse, podemos aprofundar em t√≥picos como diagn√≥stico de modelos GLM ou sele√ß√£o de vari√°veis dentro dessa estrutura.

````python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.genmod.families.family as smf_family

# Definir a semente para reprodutibilidade
np.random.seed(1)

# Criar o DataFrame similar ao "tibble" do R
df = pd.DataFrame({
    'face': np.repeat([0, 1], 25)  # 25 zeros e 25 uns, assim como rep(c(0,1), each = 25)
})

# Calcular lograte: 1 - 0.5 * face
df['lograte'] = 1 - 0.5 * df['face']

# Gerar num_comments a partir de uma distribui√ß√£o de Poisson com lambda = exp(lograte)
df['num_comments'] = np.random.poisson(lam=np.exp(df['lograte']))

# Visualizar as primeiras linhas do DataFrame (opcional)
print(df.head())

# Ajustar um modelo GLM Poisson: num_comments ~ face
model = smf.glm(formula="num_comments ~ face", data=df, family=smf_family.Poisson()).fit()

# Exibir o resumo do modelo
print(model.summary())
````

## Analises de Residuos

**O que √© a an√°lise de res√≠duos?**

A an√°lise de res√≠duos √© uma t√©cnica utilizada para avaliar a qualidade de ajuste de um modelo estat√≠stico ou de aprendizado de m√°quina. O res√≠duo √© a diferen√ßa entre o valor observado $ùë¶_ùëñ$ e o valor previsto pelo modelo $\hatùë¶_ùëñ$.

$$ùëí_ùëñ = ùë¶_ùëñ ‚àí \hatùë¶_ùëñ$$

Uma boa an√°lise de res√≠duos pode indicar:

- Se o modelo est√° bem ajustado (aus√™ncia de padr√µes nos res√≠duos).

- Poss√≠veis problemas, como heterocedasticidade ou n√£o-linearidade.

- A necessidade de ajustar o modelo ou os dados.

---

**Como interpretar os res√≠duos?**

1) Distribui√ß√£o dos res√≠duos: Os res√≠duos devem se aproximar de uma distribui√ß√£o normal com m√©dia zero.

2) Gr√°fico de res√≠duos vs. valores previstos: Deve mostrar uma distribui√ß√£o aleat√≥ria, sem padr√µes evidentes.

3) Verifica√ß√£o de outliers: Res√≠duos muito altos (positivos ou negativos) podem indicar outliers.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error #,root_mean_squared_error

# Gerando um dataset simples
np.random.seed(42)
X = np.random.rand(100, 1) * 10  # Valores independentes (entre 0 e 10)
y = 2.5 * X + np.random.randn(100, 1) * 2  # Valores dependentes com ru√≠do
y[0] = y[0] + 100  # Adicionando um outlier

# Dividindo os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Treinando um modelo de regress√£o linear
model = LinearRegression()
model.fit(X_train, y_train)

# Fazendo previs√µes no conjunto de teste
y_pred = model.predict(X_test)

# Calculando os res√≠duos
residuos = y_test - y_pred

# Gr√°fico de res√≠duos vs valores previstos
plt.scatter(y_pred, residuos, alpha=0.7)
plt.axhline(y=0, color='r', linestyle='--', linewidth=2)
plt.xlabel('Valores Previstos')
plt.ylabel('Res√≠duos')
plt.title('Gr√°fico de Res√≠duos vs. Valores Previstos')
plt.show()

# Histograma dos res√≠duos
plt.hist(residuos, bins=20, edgecolor='k', alpha=0.7)
plt.title('Distribui√ß√£o dos Res√≠duos')
plt.xlabel('Res√≠duo')
plt.ylabel('Frequ√™ncia')
plt.show()

# Histograma dos res√≠duos
plt.scatter(X, y, alpha=0.7)
plt.plot(X,model.predict(X),color='red', linewidth=2, label='Linha prevista')
plt.title('X e Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

adjusted_r2 = 1 - (1 - r2_score(y_test, y_pred)) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)


# Estat√≠sticas b√°sicas dos res√≠duos
print("M√©dia dos res√≠duos:", np.mean(residuos))
print("MSE (Erro quadr√°tico m√©dio):", mean_squared_error(y_test, y_pred))
# print("RMSE (Raiz do erro quadr√°tico m√©dio):", root_mean_squared_error(y_test, y_pred))
print("MAE (Erro absoluto m√©dio):", mean_absolute_error(y_test, y_pred))
print("MAPE (Erro percentual absoluto m√©dio):", mean_absolute_percentage_error(y_test, y_pred))
print("r2 (R2):", r2_score(y_test, y_pred))
print("R2 ajustado:", adjusted_r2)
print("X1: ", model.coef_[0], "X0: ", model.intercept_[0])
````

````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV, StratifiedKFold, KFold
from sklearn.linear_model import LinearRegression
from lightgbm import LGBMClassifier, LGBMRegressor
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PolynomialFeatures, RobustScaler
from sklearn.decomposition import PCA

from sklearn.metrics import (
    r2_score, root_mean_squared_error, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
)
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import uniform, loguniform, randint, norm
````

````python
Q1 = df.AveOccup.quantile(0.25)
Q3 = df.AveOccup.quantile(0.75)
IQR = Q3 - Q1

df = df[~((df.AveOccup < (Q1 - 1.5 * IQR)) |(df.AveOccup > (Q3 + 1.5 * IQR)))].copy()
````

````python
df['target_log'] = np.log(df['target'])
df['Population_log'] = np.log(df['Population'])
df.drop(columns=['AveBedrms'], inplace=True)
df.head()
````

````python


# # Selecionar as colunas relevantes
target = 'target_log'
ignore_cols = ['target', 'Population']
numeric_preds = list(set(df._get_numeric_data().columns) - set([target] + ignore_cols))
cat_preds = list(set(df.columns)-set(numeric_preds)-set([target] + ignore_cols))
features = list(set(numeric_preds + cat_preds) - set(ignore_cols))


# Dividir os dados em treino e teste
X = df[features]
y = df[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criar o pipeline de pr√©-processamento
numeric_transformer = RobustScaler()
poli = PolynomialFeatures(degree=3, include_bias=False)
categorical_transformer = OneHotEncoder()
pca = PCA(n_components=2)

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_preds),
        ('cat', categorical_transformer, cat_preds)
    ]
)
````

````python
# Criar o pipeline com o modelo

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('poly', poli),
    ('classifier', LinearRegression())
])

# Otimiza√ß√£o de hiperpar√¢metros com GridSearchCV
pipeline.fit(X_train, y_train)
display(pipeline)

y_pred = pipeline.predict(X_test)

y_test_exp = np.exp(y_test)
y_pred_expf = np.exp(y_pred)
r2_scoref = r2_score(y_test, y_pred)
root_mean_squared_errorf= root_mean_squared_error(y_test_exp, y_pred_expf)
mean_squared_errorf= mean_squared_error(y_test_exp, y_pred_expf)
mean_absolute_errorf = mean_absolute_error(y_test_exp, y_pred_expf)
mean_absolute_percentage_errorf= mean_absolute_percentage_error(y_test_exp, y_pred_expf)
Adj_r2 = 1 - (1-r2_score(y_test, y_pred)) * (len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)

print(f"RMSE: {root_mean_squared_errorf:.2f}")
print(f"MSE: {mean_squared_errorf:.2f}")
print(f"MAE: {mean_absolute_errorf:.2f}")
print(f"MAPE: {mean_absolute_percentage_errorf:.2f}")
print(f"R2: {r2_scoref:.2f}")
print(f"Adj R2: {Adj_r2:.2f}")

# Calculando os res√≠duos
residuos = y_test_exp - y_pred_expf

# Gr√°fico de res√≠duos vs valores previstos
plt.scatter(y_pred_expf, residuos, alpha=0.7)
plt.axhline(y=0, color='r', linestyle='--', linewidth=2)
plt.xlabel('Valores Previstos')
plt.ylabel('Res√≠duos')
plt.title('Gr√°fico de Res√≠duos vs. Valores Previstos')
plt.show()

# Histograma dos res√≠duos
plt.hist(residuos, bins=20, edgecolor='k', alpha=0.7)
plt.title('Distribui√ß√£o dos Res√≠duos')
plt.xlabel('Res√≠duo')
plt.ylabel('Frequ√™ncia')
plt.show()


````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, root_mean_squared_error, mean_absolute_percentage_error

# Gerando um dataset simples
np.random.seed(42)
X = np.random.rand(100, 1) * 10  # Valores independentes (entre 0 e 10)
y = 2.5 * X + np.random.randn(100, 1) * 2  # Valores dependentes com ru√≠do

# Dividindo os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Treinando um modelo de regress√£o linear
model = LinearRegression()
model.fit(X_train, y_train)

# Fazendo previs√µes no conjunto de teste
y_pred = model.predict(X_test)

# Calculando os res√≠duos
residuos = y_test - y_pred

# Gr√°fico de res√≠duos vs valores previstos
plt.scatter(y_pred, residuos, alpha=0.7)
plt.axhline(y=0, color='r', linestyle='--', linewidth=2)
plt.xlabel('Valores Previstos')
plt.ylabel('Res√≠duos')
plt.title('Gr√°fico de Res√≠duos vs. Valores Previstos')
plt.show()

# Histograma dos res√≠duos
plt.hist(residuos, bins=20, edgecolor='k', alpha=0.7)
plt.title('Distribui√ß√£o dos Res√≠duos')
plt.xlabel('Res√≠duo')
plt.ylabel('Frequ√™ncia')
plt.show()

# Histograma dos res√≠duos
plt.scatter(X, y, alpha=0.7)
plt.plot(X,model.predict(X),color='red', linewidth=2, label='Linha prevista')
plt.title('X e Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

adjusted_r2 = 1 - (1 - r2_score(y_test, y_pred)) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)


# Estat√≠sticas b√°sicas dos res√≠duos
print("M√©dia dos res√≠duos:", np.mean(residuos))
print("Mediana dos res√≠duos:", np.median(residuos))
print("MSE (Erro quadr√°tico m√©dio):", mean_squared_error(y_test, y_pred))
print("RMSE (Raiz do erro quadr√°tico m√©dio):", root_mean_squared_error(y_test, y_pred))
print("MAE (Erro absoluto m√©dio):", mean_absolute_error(y_test, y_pred))
print("MAPE (Erro percentual absoluto m√©dio):", mean_absolute_percentage_error(y_test, y_pred))
print("r2 (R2):", r2_score(y_test, y_pred))
print("R2 ajustado:", adjusted_r2)
print("X1: ", model.coef_, "X0: ", model.intercept_)
````
