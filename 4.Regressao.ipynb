# Regressão

## Metricas de Avaliação

$R^2$, $R^2$ ajustado, MSE, RMSE, MAE, MAPE

---

**Coeficiente de Determinação (R²)**

O **R²**, ou coeficiente de determinação, é uma medida que indica **a proporção da variabilidade dos dados dependentes que é explicada pelo modelo de regressão**. Em outras palavras, ele nos diz o quanto das variações em \( Y \) (variável resposta) é capturado pelas variações em \( X \) (variáveis explicativas).

**### Como é Calculado?**

Para um modelo de regressão, o R² é calculado com a fórmula:

$$
R^2 = 1 - \frac{SSE}{SST} = 1 - \frac{\sum_{i = 1}^n (y_i - \hat{y_i})^2  }{\sum_{i = 1}^n (y_i - \bar{y})^2}
$$

Onde:
- **SSE (Soma dos Quadrados dos Erros)**: É a soma dos quadrados das diferenças entre os valores observados e os valores preditos pelo modelo.
- **SST (Soma Total dos Quadrados)**: É a soma dos quadrados das diferenças entre os valores observados e a média dos valores observados.

Um **R²** de 0,8, por exemplo, significa que 80% da variação total em \( Y \) é explicada por \( X \) segundo o modelo.

**### Interpretação**

- **Variação Explicada**: Valores próximos de 1 indicam que o modelo explica grande parte da variabilidade dos dados.
- **Limitações**:  
  - **Cresce com a inclusão de variáveis**: Mesmo que elas não tenham um relacionamento significativo com a variável dependente.
  - **Não diz se o modelo é apropriado**: Um R² alto não garante que os pressupostos da regressão foram atendidos nem que não haja viés.

---

**R² Ajustado**

O **R² Ajustado** foi desenvolvido para lidar com uma limitação importante do R²: sua tendência a aumentar com a adição de novas variáveis, mesmo quando estas não contribuem significativamente para o modelo.

**### Fórmula do R² Ajustado**

$$
R^2_{adj} = 1 - \left( \frac{(1-R^2)(n-1)}{n - k - 1} \right)
$$

Onde:
- **\( n \)**: Número total de observações.
- **\( k \)**: Número de variáveis explicativas (excluindo o intercepto).

**### Por Que Usá-lo?**

- **Penalização de Variáveis Inúteis**: O R² Ajustado diminui quando adicionamos uma variável que não melhora realmente o modelo. Assim, ele favorece modelos mais parcimoniosos.
- **Comparação Entre Modelos**: É mais adequado quando queremos comparar modelos com diferentes números de preditores, pois ele equilibra a qualidade do ajuste e a complexidade do modelo.

**### Exemplo Ilustrativo**

Imagine dois modelos para prever o preço de imóveis:

- **Modelo A**: Usa as variáveis área, número de quartos e idade do imóvel.
- **Modelo B**: Usa as mesmas variáveis do Modelo A, plus a cor da porta.

Suponhamos que o Modelo B apresente um R² ligeiramente maior que o A, mas essa variável extra (cor da porta) não possui um real impacto prático. O **R² Ajustado** do Modelo B poderá ser menor que a do Modelo A, indicando que o acréscimo não melhorou a capacidade preditiva do modelo de forma significativa.




**Conclusão**

- **R²** mostra a fração da variabilidade de \( Y \) explicada pelo modelo e fornece uma visão geral do quão bem os dados se ajustam à regressão.
- **R² Ajustado** aprimora essa medida, penalizando a inclusão de preditores que não agregam valor real, permitindo uma comparação mais justa entre modelos com diferentes quantidades de variáveis.

1. Erro Quadrático Médio (MSE - Mean Squared Error)

    O **MSE** é a média dos quadrados dos erros (diferença entre valores reais e preditos):

    $$
    \text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
    $$

    **Principais pontos**:

    - **Sensibilidade a Outliers**: Como os erros são elevados ao quadrado, erros grandes têm impacto muito mais pronunciado.
    - **Unidade**: Está na unidade ao quadrado da variável alvo, o que pode dificultar interpretações diretas.
    - **Uso em Otimização**: É amplamente utilizada em algoritmos de otimização por ser diferenciável, facilitando métodos como gradiente descendente.



2. Raiz do Erro Quadrático Médio (RMSE - Root Mean Squared Error)

    O **RMSE** é a raiz quadrada do MSE:

    $$
    \text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2} = \sqrt{\text{MSE}}
    $$

    **Principais pontos**:

    - **Interpretação em Unidades Originais**: Como o RMSE retira a raiz quadrada, ele volta para as mesmas unidades dos dados, facilitando a comparação com os valores reais.
    - **Sensibilidade a Outliers**: Assim como o MSE, continua a punir erros grandes, pois está diretamente ligado aos quadrados dos erros.
    - **Comparabilidade**: Serve para avaliar a dispersão dos erros e, em muitos casos, é comparado com a variabilidade natural dos dados.



3. Erro Absoluto Médio (MAE - Mean Absolute Error)

    O **MAE** calcula a média das diferenças absolutas:

    $$
    \text{MAE} = \frac{1}{n} \sum_{i=1}^{n} \left| y_i - \hat{y}_i \right|
    $$

    **Principais pontos**:
    - **Robustez a Outliers**: Ao utilizar o valor absoluto, cada erro contribui de forma linear. Isso o torna menos sensível a outliers extremos comparado às métricas baseadas em quadrados.
    - **Unidade**: Está na mesma unidade dos valores de \( y \), facilitando a interpretação.
    - **Intuição**: O MAE representa, em média, o quanto as previsões se afastam dos valores reais.



4. Erro Percentual Absoluto Médio (MAPE - Mean Absolute Percentage Error)

    O **MAPE** expressa o erro como uma porcentagem do valor real:

    $$
    \text{MAPE} = \frac{100\%}{n} \sum_{i=1}^{n} \left| \frac{y_i - \hat{y}_i}{y_i} \right|
    $$

    **Principais pontos**:
    - **Interpretação Relativa**: Por trazer o erro em termos percentuais, o MAPE permite comparar a performance do modelo em escalas diferentes.
    - **Problemas com Valores Próximos de Zero**: Se \( $y_i$ \) for zero (ou muito próximo), o MAPE pode tender ao infinito ou distorcer a média, exigindo cuidados.
    - **Aplicabilidade**: Ideal quando se deseja saber o quanto, em termos relativos, as previsões se desviam dos valores reais, especialmente em contextos como previsão de vendas, onde percentuais são altamente informativos.

**Tabela Resumo das Métricas**

| Métrica | Fórmula | Sensibilidade a Outliers | Unidade | Interpretação |
|---------|---------|--------------------------|---------|---------------|
| **MSE**   | $\frac{1}{n} \sum (y - \hat{y})^2$ | Alta (eleva os erros ao quadrado)       | Ao quadrado de y | Média dos erros ao quadrado |
| **RMSE**  | $\sqrt{\text{MSE}}$               | Alta (transmite a sensibilidade do MSE) | Mesma que y   | Erro médio na mesma unidade de y |
| **MAE**   | $\frac{1}{n} \sum \|y - \hat{y}\|$   | Moderada (linear)                       | Mesma que y   | Erro absoluto médio |
| **MAPE**  | $\frac{100\%}{n} \sum \left\|\frac{y - \hat{y}}{y}\right\|$ | Pode ser distorcido para y próximo de zero | Percentual | Porcentagem média de erro em relação aos valores reais |

---

**Considerações Finais**

- **Escolha da Métrica**:  
  - Se seu projeto tem baixa tolerância a grandes erros, o RMSE pode ser mais adequado, pois penaliza desvios importantes.
  - Se você quer uma interpretação intuitiva e menos sensível a outliers, o MAE pode ser mais indicado.
  - Para comparações em contextos onde a escala dos dados varia (ou quando percentuais são mais relevantes), o MAPE é bastante útil – **mas esteja atento a valores pequenos no denominador**.

- **Interpretação Prática**:  
  Imagine que você esteja prevendo os preços de imóveis. Um RMSE de R\$50.000 pode indicar que, em média, a diferença entre o valor previsto e o real é de R\$50.000, especialmente se o preço médio dos imóveis estiver em centenas de milhares. Por outro lado, um MAPE de 5% diz que, em média, seu modelo erra 5% dos preços reais, uma interpretação que pode ser mais intuitiva para um investidor.

- **Implicações para Otimização e Validação**:  
  Muitas vezes, durante o treinamento do modelo, a função de custo é formulada em termos de MSE ou RMSE devido à facilidade de derivação para algoritmos de otimização. No entanto, na validação, usar MAE ou MAPE pode oferecer outra perspectiva sobre a adequação do modelo para o uso prático.



## Regularização Lasso (L1)

**Por Que Regularizar?**

Ao treinar modelos (por exemplo, modelos de regressão), nosso objetivo é capturar a relação subjacente entre as variáveis sem “aprender” o ruído presente nos dados de treino. A regularização atua adicionando uma penalização aos coeficientes do modelo. Essa penalização encoraja soluções mais simples, controlando o tamanho dos coeficientes e, em alguns casos, zerando aqueles que são menos relevantes.


### Conceito

A regularização L1 adiciona a soma dos valores absolutos dos coeficientes à função de custo. Essa abordagem incentiva muitos coeficientes a se tornarem exatamente zero, promovendo **sparsidade** e, portanto, funcionando também como um método de seleção de variáveis.

### Fórmula

Seja $ J(w) $ a função de custo com regularização incorporada:

$$
J(w) = \text{Loss}(w) + \lambda \sum_{i=1}^{n} |w_i|
$$

- **$ \text{Loss}(w) $**: Pode ser a soma dos erros quadráticos, por exemplo.
- **$ \lambda $**: É o parâmetro de regularização que controla a intensidade da penalização.
- **$ \sum_{i=1}^{n} |w_i| $**: Soma dos valores absolutos dos coeficientes.

### Efeitos e Intuição

- **Sparsidade**: Muitos coeficientes podem ser reduzidos exatamente a zero. Isso pode ser extremamente útil quando se deseja um modelo mais interpretável, focando nas variáveis que realmente importam.
- **Seleção de Variáveis**: Ao eliminar gradualmente os coeficientes irrelevantes, o modelo indica quais preditores possuem relação mais forte com a variável de interesse.
- **Robustez a Overfitting**: Ao penalizar os coeficientes, o modelo torna-se menos suscetível à variação nos dados de treino.

---


## Regularização L2 (Ridge)

### Conceito

A regularização L2 penaliza os coeficientes através da soma dos quadrados de seus valores. Diferente da L1, essa técnica **encolhe todos os coeficientes**, mas dificilmente os zera completamente.

### Fórmula

A função de custo com regularização L2 é dada por:

$$
J(w) = \text{Loss}(w) + \lambda \sum_{i=1}^{n} w_i^2
$$

- **$ \sum_{i=1}^{n} w_i^2 $**: Soma dos quadrados dos coeficientes, que enfatiza penalizações maiores para coeficientes mais extremos.

### Efeitos e Intuição

- **Redução da Variância**: Ao encolher os coeficientes, o modelo tende a ser menos sensível às pequenas variações dos dados de treino, melhorando a capacidade de generalização.
- **Multicolinearidade**: Em cenários onde as variáveis preditoras são altamente correlacionadas, o L2 distribui a penalização entre elas, estabilizando a estimativa dos coeficientes.
- **Não Esparso**: Diferentemente da L1, a Ridge preserva todas as variáveis, o que pode ser desejável se acreditarmos que todas têm alguma relevância.

| Aspecto                  | Regularização L1 (Lasso)                                         | Regularização L2 (Ridge)                                        |
|--------------------------|------------------------------------------------------------------|-----------------------------------------------------------------|
| **Penalização**          | Soma dos valores absolutos dos coeficientes<br>$\sum \|w_i\|$    | Soma dos quadrados dos coeficientes<br>$\sum w_i^2$            |

| **Seleção de Variáveis** | Funciona como uma forma de seleção de variáveis                    | Mantém todas as variáveis no modelo                              |
| **Aplicação**            | Útil quando se acredita que apenas algumas variáveis importam      | Útil quando todas as variáveis possuem alguma relevância         |
| **Robustez**             | Pode criar modelos mais simples e interpretáveis                   | Reduz variações e instabilidades especialmente sob multicolinearidade |

## Elastic Net

- **Parâmetro $ \lambda $**: Em ambos os casos, a escolha do $ \lambda $ é crucial. Valores muito altos podem simplificar demais o modelo (subajuste), enquanto valores muito baixos podem não reduzir suficientemente o overfitting.
- **Elastic Net**: Em alguns cenários, é interessante combinar os efeitos de L1 e L2. O **Elastic Net** integra as duas técnicas:

  $$
  J(w) = \text{Loss}(w) + \lambda_1 \sum_{i=1}^{n} |w_i| + \lambda_2 \sum_{i=1}^{n} w_i^2
  $$

  Essa combinação aproveita a capacidade do L1 de eliminar variáveis irrelevantes e a estabilidade e suavização do L2.

---

### Reflexões Finais

A escolha entre L1 e L2 (ou uma combinação por meio do Elastic Net) depende do contexto do problema:

- Se você deseja um modelo mais **interpretável** com menos variáveis, a **regularização L1** pode ser o caminho.
- Se a intenção é apenas reduzir a variância e melhorar a generalização, mantendo todas as variáveis, a **regularização L2** é apropriada.

## Árvore de regressão 

---

### 1. Conceito Básico

Uma árvore de decisão para regressão é um modelo _não paramétrico_ que divide o espaço de preditores em regiões homogêneas, de modo que, para cada região, a predição é feita através de um valor constante – geralmente a média dos valores de $ y $ dos dados que caem naquela região. Esse método é especialmente útil para capturar relações não lineares entre as variáveis.

---

### 2. Processo de Divisão (Splitting)

O algoritmo constrói a árvore de forma recursiva. Em cada nó, ele avalia todas as possíveis divisões (ou cortes) em cada variável preditora. O critério utilizado geralmente é a minimização do erro quadrático. Em termos matemáticos, se dividirmos um nó em duas regiões, $ R_1 $ e $ R_2 $, queremos que a soma dos erros quadráticos em cada região seja a menor possível:

$$
\text{Erro Total} = \sum_{i \in R_1} \left( y_i - \bar{y}_{R_1} \right)^2 + \sum_{i \in R_2} \left( y_i - \bar{y}_{R_2} \right)^2
$$

Onde:
- $ \bar{y}_{R_1} $ é a média dos valores de $ y $ na região $ R_1 $.
- $ \bar{y}_{R_2} $ é a média dos valores de $ y $ na região $ R_2 $.

O algoritmo procura o par (variável, valor-limiar) que minimiza essa soma. Essa divisão é realizada recursivamente em cada nó até que alguma condição de parada seja atendida.

---

### 3. Critérios de Parada e Poda

Para evitar **overfitting**, a árvore geralmente não é construída até que cada nó contenha apenas um exemplo. Alguns dos critérios de parada incluem:

- **Profundidade máxima:** Limitar a profundidade da árvore.
- **Número mínimo de amostras em um nó:** Exigir que cada nó contenha pelo menos um certo número de exemplos.
- **Ganho mínimo na redução do erro:** Parar se uma divisão não resultar em uma redução significativa na soma dos erros quadráticos.

Além disso, pode ser aplicada a **poda da árvore**, que consiste em remover nós que pouco contribuem para a predição, melhorando a generalização do modelo.

---

### 4. Estrutura da Árvore e Predição

Cada nó interno da árvore representa uma condição do tipo:

$$
x_j < t
$$

- $ x_j $ é uma variável preditora.
- $ t $ é o limiar escolhido para a divisão.

A árvore se ramifica de acordo com essas condições, e cada região final (ou folha) terá uma previsão constante, tipicamente a média dos $ y $ dos exemplos naquela folha:

$$
\hat{y} = \bar{y}_{R}
$$

Visualmente, a estrutura da árvore pode ser ilustrada da seguinte forma:

```
                     [Nó Raiz]
                         │
                  (Regra: X1 < 2.5)
                      /       \
           [Nó Esquerdo]   [Nó Direito]
                |                |
       (X2 < 1.7)           (X3 < 0.8)
        /         \          /         \
  Folha 1      Folha 2   Folha 3       Folha 4
   (média)      (média)    (média)      (média)
```

---

### 5. Vantagens e Desvantagens

#### Vantagens

- **Interpretação:**  
  Fácil de visualizar e entender, pois as regras de decisão são claras.

- **Captura de Não-linearidades:**  
  Boa capacidade de modelar relações não lineares entre as variáveis.

- **Sem Necessidade de Escalonamento:**  
  Não exige normalização ou padronização dos preditores.

#### Desvantagens

- **Tendência ao Overfitting:**  
  Árvores muito profundas podem ajustar demais os dados de treinamento. O uso de poda ou ensembles (como _Random Forests_ e _Gradient Boosting_) pode ajudar.

- **Alta Variância:**  
  Pequenas mudanças nos dados podem resultar em árvores bem diferentes, o que dificulta a estabilidade do modelo.

---


### Conclusão

Árvores de decisão para regressão oferecem uma abordagem intuitiva e flexível para modelar relações complexas entre variáveis. Elas segmentam o espaço de dados através de regras lógicas e geram predições simples (médias) para cada região definida. Embora sejam fáceis de interpretar, é importante usar técnicas de controle de complexidade (como poda ou ensembles) para evitar que o modelo se ajuste excessivamente aos dados de treinamento.

## Modelos Lineares Generalizados (GLM)

Os **Modelos Lineares Generalizados (GLM)** são uma extensão elegante dos modelos de regressão linear clássicos. Eles permitem modelar situações onde a variável resposta não segue necessariamente uma distribuição normal, adaptando-se a diversos tipos de dados e relações. Vamos explorar os principais componentes e ideias por trás dos GLMs.

Um GLM é composto por três elementos fundamentais:

1. **Componente Aleatório (Distribuição da Resposta):**  

   Aqui, assume-se que a variável resposta $Y$ possui uma distribuição pertencente à família exponencial. Essa família inclui distribuições como a normal, binomial, Poisson, gama, entre outras. Essa flexibilidade permite modelar:
   - Dados contínuos com ruídos gaussianos (regressão linear).
   - Dados binários (regressão logística).
   - Dados de contagem (regressão Poisson).

2. **Componente Sistemático (Preditor Linear):**  

   É a parte que incorpora as variáveis explicativas ou preditores. Nela, usamos uma combinação linear dos preditores:
   
   $$
   \eta = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_k x_k
   $$
   
   onde $\eta$ é chamado de **predictor linear** e os coeficientes $\beta_i$ são os parâmetros a serem estimados.

3. **Função de Ligação (Link Function):**  

   Essa função conecta o preditor linear à média condicional da variável resposta. Ou seja, define uma relação entre $\mu = E(Y)$ e o preditor $\eta$:
   
   $$
   g(\mu)= \eta
   $$
   
   A escolha da função de ligação $g(\cdot)$ depende da natureza dos dados. Por exemplo:
   
   - **Identidade:** Para dados gaussianos, onde $g(\mu)=\mu$.
   - **Logito:** Para dados binários, usando $g(\mu)=\log\left(\frac{\mu}{1-\mu}\right)$.
   - **Logaritmo:** Para contagens (regressão Poisson), onde $g(\mu)=\log(\mu)$.

---


### Estrutura Matemática de um GLM

Em resumo, o modelo assume que cada observação $y_i$ tem uma distribuição da família exponencial, com a função de densidade ou massa escrita genericamente como

$$
f(y_i; \theta_i, \phi) = \exp\left\{\frac{y_i \theta_i - b(\theta_i)}{\phi} + c(y_i,\phi)\right\},
$$

onde:
- $\theta_i$ é o parâmetro natural (ou canônico) do modelo, que, em muitos casos, está diretamente relacionado à média.
- $\phi$  é o parâmetro de dispersão (ou escala), presente em muitas distribuições para ajustar a variabilidade dos dados.
- $b(\theta)$ é uma função conhecida como cumulante, que garante a normalização da função.
- $c(y,\phi)$ é uma função que depende somente de $y$ e $\phi$ e, em geral, serve para normalizar a distribuição.


$$
f_X(x|\theta) = h(x)\exp[\eta(\theta).T(x) - A(\theta)]
$$

A função de ligação é então usada para relacionar o parâmetro de localização (muitas vezes, a média $\mu_i$) ao preditor linear, através de

$$
g(\mu_i) = \beta_0 + \beta_1 x_{i1} + \cdots + \beta_k x_{ik}.
$$


### Valor Esperado

O valor esperado (média) de $y_i$ é dado por:

$$
E(y_i) = \mu_i = b'(\theta_i)
$$

Ou seja, a média dos dados está diretamente relacionada à primeira derivada de $b(\theta)$ em relação a $\theta$.

### Variância

A variância de $y_i$ é expressa como:

$$
\operatorname{Var}(y_i) = \phi \, b''(\theta_i)
$$

Isso mostra que a variância é proporcional à segunda derivada de $b(\theta)$ e ajustada pelo parâmetro de dispersão $\phi$.


### Comparativos

| Tipo de GLM            | Distribuição da Resposta | Função de Ligação       | Aplicação Comum                             |
|------------------------|--------------------------|-------------------------|---------------------------------------------|
| **Regressão Linear**   | Normal                   | Identidade: $$g(\mu) = \mu$$ | Previsão de valores contínuos               |
| **Regressão Logística**| Binomial                 | Logito: $$g(\mu) = \log\left(\frac{\mu}{1-\mu}\right)$$ | Modelagem de probabilidades (classificação) |
| **Regressão Poisson**  | Poisson                  | Logaritmo: $$g(\mu) = \log(\mu)$$ | Contagem de eventos (ex.: número de ocorrências) |

---

### Por Que Utilizar GLMs?

- **Flexibilidade:** Permite modelar diferentes tipos de variáveis resposta (contínuas, binárias, de contagem, etc.).
- **Integração com a Família Exponencial:** A abordagem unifica modelos diferentes sob uma mesma estrutura, facilitando técnicas de estimação, diagnóstico e validação.
- **Interpretação:** Embora a interpretação dos coeficientes dependa da função de ligação escolhida, os GLMs fornecem uma forma consistente de analisar a relação entre preditores e resposta.

---

### Reflexão e Aplicações

Pense nos GLMs como uma poderosa caixa de ferramentas para lidar com uma variedade de problemas de modelagem. Por exemplo, se você estiver trabalhando com dados de incidência de uma doença em diferentes regiões, a regressão Poisson pode ser o modelo ideal. Ou, se a questão for prever o sucesso ou fracasso de uma campanha, a regressão logística se encaixa perfeitamente.

## Analises de Residuos

**O que é a análise de resíduos?**

A análise de resíduos é uma técnica utilizada para avaliar a qualidade de ajuste de um modelo estatístico ou de aprendizado de máquina. O resíduo é a diferença entre o valor observado $𝑦_𝑖$ e o valor previsto pelo modelo $\hat𝑦_𝑖$.

$$𝑒_𝑖 = 𝑦_𝑖 − \hat𝑦_𝑖$$

Uma boa análise de resíduos pode indicar:

- Se o modelo está bem ajustado (ausência de padrões nos resíduos).

- Possíveis problemas, como heterocedasticidade ou não-linearidade.

- A necessidade de ajustar o modelo ou os dados.

---

**Como interpretar os resíduos?**

1) Distribuição dos resíduos: Os resíduos devem se aproximar de uma distribuição normal com média zero.

2) Gráfico de resíduos vs. valores previstos: Deve mostrar uma distribuição aleatória, sem padrões evidentes.

3) Verificação de outliers: Resíduos muito altos (positivos ou negativos) podem indicar outliers.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error #,root_mean_squared_error

# Gerando um dataset simples
np.random.seed(42)
X = np.random.rand(100, 1) * 10  # Valores independentes (entre 0 e 10)
y = 2.5 * X + np.random.randn(100, 1) * 2  # Valores dependentes com ruído
y[0] = y[0] + 100  # Adicionando um outlier

# Dividindo os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Treinando um modelo de regressão linear
model = LinearRegression()
model.fit(X_train, y_train)

# Fazendo previsões no conjunto de teste
y_pred = model.predict(X_test)

# Calculando os resíduos
residuos = y_test - y_pred

# Gráfico de resíduos vs valores previstos
plt.scatter(y_pred, residuos, alpha=0.7)
plt.axhline(y=0, color='r', linestyle='--', linewidth=2)
plt.xlabel('Valores Previstos')
plt.ylabel('Resíduos')
plt.title('Gráfico de Resíduos vs. Valores Previstos')
plt.show()

# Histograma dos resíduos
plt.hist(residuos, bins=20, edgecolor='k', alpha=0.7)
plt.title('Distribuição dos Resíduos')
plt.xlabel('Resíduo')
plt.ylabel('Frequência')
plt.show()

# Histograma dos resíduos
plt.scatter(X, y, alpha=0.7)
plt.plot(X,model.predict(X),color='red', linewidth=2, label='Linha prevista')
plt.title('X e Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

adjusted_r2 = 1 - (1 - r2_score(y_test, y_pred)) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)


# Estatísticas básicas dos resíduos
print("Média dos resíduos:", np.mean(residuos))
print("MSE (Erro quadrático médio):", mean_squared_error(y_test, y_pred))
# print("RMSE (Raiz do erro quadrático médio):", root_mean_squared_error(y_test, y_pred))
print("MAE (Erro absoluto médio):", mean_absolute_error(y_test, y_pred))
print("MAPE (Erro percentual absoluto médio):", mean_absolute_percentage_error(y_test, y_pred))
print("r2 (R2):", r2_score(y_test, y_pred))
print("R2 ajustado:", adjusted_r2)
print("X1: ", model.coef_[0], "X0: ", model.intercept_[0])
````

````python
df['target_log'] = np.log(df['target'])
df['Population_log'] = np.log(df['Population'])
df.drop(columns=['AveBedrms'], inplace=True)
df.head()
````

````python


# # Selecionar as colunas relevantes
target = 'target_log'
ignore_cols = ['target', 'Population']
numeric_preds = list(set(df._get_numeric_data().columns) - set([target] + ignore_cols))
cat_preds = list(set(df.columns)-set(numeric_preds)-set([target] + ignore_cols))
features = list(set(numeric_preds + cat_preds) - set(ignore_cols))


# Dividir os dados em treino e teste
X = df[features]
y = df[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criar o pipeline de pré-processamento
numeric_transformer = RobustScaler()
poli = PolynomialFeatures(degree=3, include_bias=False)
categorical_transformer = OneHotEncoder()
pca = PCA(n_components=2)

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_preds),
        ('cat', categorical_transformer, cat_preds)
    ]
)
````

````python
# Criar o pipeline com o modelo

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('poly', poli),
    ('classifier', LinearRegression())
])

# Otimização de hiperparâmetros com GridSearchCV
pipeline.fit(X_train, y_train)
display(pipeline)

y_pred = pipeline.predict(X_test)

y_test_exp = np.exp(y_test)
y_pred_expf = np.exp(y_pred)
r2_scoref = r2_score(y_test, y_pred)
root_mean_squared_errorf= root_mean_squared_error(y_test_exp, y_pred_expf)
mean_squared_errorf= mean_squared_error(y_test_exp, y_pred_expf)
mean_absolute_errorf = mean_absolute_error(y_test_exp, y_pred_expf)
mean_absolute_percentage_errorf= mean_absolute_percentage_error(y_test_exp, y_pred_expf)
Adj_r2 = 1 - (1-r2_score(y_test, y_pred)) * (len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)

print(f"RMSE: {root_mean_squared_errorf:.2f}")
print(f"MSE: {mean_squared_errorf:.2f}")
print(f"MAE: {mean_absolute_errorf:.2f}")
print(f"MAPE: {mean_absolute_percentage_errorf:.2f}")
print(f"R2: {r2_scoref:.2f}")
print(f"Adj R2: {Adj_r2:.2f}")

# Calculando os resíduos
residuos = y_test_exp - y_pred_expf

# Gráfico de resíduos vs valores previstos
plt.scatter(y_pred_expf, residuos, alpha=0.7)
plt.axhline(y=0, color='r', linestyle='--', linewidth=2)
plt.xlabel('Valores Previstos')
plt.ylabel('Resíduos')
plt.title('Gráfico de Resíduos vs. Valores Previstos')
plt.show()

# Histograma dos resíduos
plt.hist(residuos, bins=20, edgecolor='k', alpha=0.7)
plt.title('Distribuição dos Resíduos')
plt.xlabel('Resíduo')
plt.ylabel('Frequência')
plt.show()


````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, root_mean_squared_error, mean_absolute_percentage_error

# Gerando um dataset simples
np.random.seed(42)
X = np.random.rand(100, 1) * 10  # Valores independentes (entre 0 e 10)
y = 2.5 * X + np.random.randn(100, 1) * 2  # Valores dependentes com ruído

# Dividindo os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Treinando um modelo de regressão linear
model = LinearRegression()
model.fit(X_train, y_train)

# Fazendo previsões no conjunto de teste
y_pred = model.predict(X_test)

# Calculando os resíduos
residuos = y_test - y_pred

# Gráfico de resíduos vs valores previstos
plt.scatter(y_pred, residuos, alpha=0.7)
plt.axhline(y=0, color='r', linestyle='--', linewidth=2)
plt.xlabel('Valores Previstos')
plt.ylabel('Resíduos')
plt.title('Gráfico de Resíduos vs. Valores Previstos')
plt.show()

# Histograma dos resíduos
plt.hist(residuos, bins=20, edgecolor='k', alpha=0.7)
plt.title('Distribuição dos Resíduos')
plt.xlabel('Resíduo')
plt.ylabel('Frequência')
plt.show()

# Histograma dos resíduos
plt.scatter(X, y, alpha=0.7)
plt.plot(X,model.predict(X),color='red', linewidth=2, label='Linha prevista')
plt.title('X e Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

adjusted_r2 = 1 - (1 - r2_score(y_test, y_pred)) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)


# Estatísticas básicas dos resíduos
print("Média dos resíduos:", np.mean(residuos))
print("Mediana dos resíduos:", np.median(residuos))
print("MSE (Erro quadrático médio):", mean_squared_error(y_test, y_pred))
print("RMSE (Raiz do erro quadrático médio):", root_mean_squared_error(y_test, y_pred))
print("MAE (Erro absoluto médio):", mean_absolute_error(y_test, y_pred))
print("MAPE (Erro percentual absoluto médio):", mean_absolute_percentage_error(y_test, y_pred))
print("r2 (R2):", r2_score(y_test, y_pred))
print("R2 ajustado:", adjusted_r2)
print("X1: ", model.coef_, "X0: ", model.intercept_)
````
